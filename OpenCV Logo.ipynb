```python
import cv2
import numpy as np
```

```python
image=np.ones((800,600,3),np.uint8)
image.fill(255)

#red color
image = cv2.ellipse(image,(300,200),(80,80),0,0,50,(0,0,255),50)
image = cv2.ellipse(image,(300,200),(80,80),0,130,360,(0,0,255),50)

#green color
image = cv2.ellipse(image,(190,400),(80,80),0,-10,270,(0,255,0),50)


#blue color
image = cv2.ellipse(image,(410,400),(80,80),0,-50,230,(255,0,0),50)
# image = cv2.ellipse(image,(410,400),(80,80),0,255,360,(255,0,0),50)


font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(image,'OpenCV',(60,620), font,4,(0,0,0),10,cv2.LINE_AA)
cv2.imwrite('OpenCV_Icon.png',image)

cv2.imshow('OpenCV',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
